# How to classify "wine" using sklearn linear_models - Multiclass Classification

```python
## How to classify "wine" using sklearn linear_models - Multiclass Classification
## DataSet: skleran.datasets.load_wine()
def Snippet_172(): 
    print()
    print(format('How to classify "wine" using sklearn linear_models','*^82'))    
    
    import warnings
    warnings.filterwarnings("ignore")
    
    # load libraries
    from sklearn import datasets
    from sklearn import metrics
    from sklearn import linear_model
    from sklearn.model_selection import train_test_split
    import matplotlib.pyplot as plt    
    
    plt.style.use('ggplot')
    
    # load the iris datasets
    dataset = datasets.load_wine()
    X = dataset.data; y = dataset.target
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
    
    # fit a linear_model.SGDClassifier() model to the data
    model = linear_model.SGDClassifier()
    model.fit(X_train, y_train)
    print(); print(model)
    
    # make predictions
    expected_y  = y_test
    predicted_y = model.predict(X_test)
    
    # summarize the fit of the model
    print(); print(metrics.classification_report(expected_y, predicted_y, 
                   target_names=dataset.target_names))
    print(); print(metrics.confusion_matrix(expected_y, predicted_y))

    # fit a linear_model.PassiveAggressiveClassifier() model to the data
    model = linear_model.PassiveAggressiveClassifier()
    model.fit(X_train, y_train)
    print(); print(model)
    
    # make predictions
    expected_y  = y_test
    predicted_y = model.predict(X_test)
    
    # summarize the fit of the model
    print(); print(metrics.classification_report(expected_y, predicted_y, 
                   target_names=dataset.target_names))
    print(); print(metrics.confusion_matrix(expected_y, predicted_y))
    
    
    # fit a linear_model.PassiveAggressiveClassifier() model to the data
    model = linear_model.RidgeClassifier()
    model.fit(X_train, y_train)
    print(); print(model)
    
    # make predictions
    expected_y  = y_test
    predicted_y = model.predict(X_test)
    
    # summarize the fit of the model
    print(); print(metrics.classification_report(expected_y, predicted_y, 
                   target_names=dataset.target_names))
    print(); print(metrics.confusion_matrix(expected_y, predicted_y))    
    
Snippet_172()
```
