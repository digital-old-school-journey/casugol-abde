# How to classify "wine" using sklearn LDA and QDA model - Multiclass Classification

```python
## How to classify "wine" using sklearn LDA and QDA model - Multiclass Classification
## DataSet: skleran.datasets.load_wine()
def Snippet_175(): 
    print()
    print(format('How to classify "wine" using sklearn LDA and QDA model - Multiclass Classification','*^82'))    
    import warnings
    warnings.filterwarnings("ignore")
    # load libraries
    from sklearn import datasets
    from sklearn import metrics
    from sklearn.model_selection import train_test_split
    import matplotlib.pyplot as plt    
    
    plt.style.use('ggplot')
    
    from sklearn import discriminant_analysis
    # load the iris datasets
    dataset = datasets.load_wine()
    X = dataset.data; y = dataset.target
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
    
    # fit a LinearDiscriminantAnalysis() model to the data
    model = discriminant_analysis.LinearDiscriminantAnalysis()
    model.fit(X_train, y_train)
    print(); print(model)
    # make predictions
    expected_y  = y_test
    predicted_y = model.predict(X_test)
    # summarize the fit of the model
    print(); print('LinearDiscriminantAnalysis(): ')
    print(); print(metrics.classification_report(expected_y, predicted_y, 
                   target_names=dataset.target_names))
    print(); print(metrics.confusion_matrix(expected_y, predicted_y))

    # fit a QuadraticDiscriminantAnalysis() model to the data
    model = discriminant_analysis.QuadraticDiscriminantAnalysis()
    model.fit(X_train, y_train)
    print(); print(model)
    # make predictions
    expected_y  = y_test
    predicted_y = model.predict(X_test)
    # summarize the fit of the model
    print(); print('QuadraticDiscriminantAnalysis(): ')
    print(); print(metrics.classification_report(expected_y, predicted_y, 
                   target_names=dataset.target_names))
    print(); print(metrics.confusion_matrix(expected_y, predicted_y))
    
    # fit a QuadraticDiscriminantAnalysis() model to the data
    model = discriminant_analysis.QuadraticDiscriminantAnalysis()
    model.fit(X_train, y_train)
    print(); print(model)
    # make predictions
    expected_y  = y_test
    predicted_y = model.predict(X_test)
    # summarize the fit of the model
    print(); print('QuadraticDiscriminantAnalysis(): ')
    print(); print(metrics.classification_report(expected_y, predicted_y, 
                   target_names=dataset.target_names))
    print(); print(metrics.confusion_matrix(expected_y, predicted_y))    
Snippet_175()
```
