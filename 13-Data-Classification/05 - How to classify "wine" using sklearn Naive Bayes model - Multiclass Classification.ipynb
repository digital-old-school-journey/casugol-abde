# How to classify "wine" using sklearn Naive Bayes mdeol - Multiclass Classification

```python
## How to classify "wine" using sklearn Naive Bayes mdeol - Multiclass Classification
## DataSet: skleran.datasets.load_wine()
def Snippet_173(): 
    print()
    print(format('How to classify "wine" using sklearn Naive Bayes model - Multiclass Classification','*^82'))    
    import warnings
    warnings.filterwarnings("ignore")
    # load libraries
    from sklearn import datasets
    from sklearn import metrics
    from sklearn.model_selection import train_test_split
    import matplotlib.pyplot as plt    
    
    plt.style.use('ggplot')

    from sklearn import naive_bayes
    # load the iris datasets
    dataset = datasets.load_wine()
    X = dataset.data; y = dataset.target
    
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
    
    # fit a naive_bayes.BernoulliNB() model to the data
    model = naive_bayes.BernoulliNB()
    model.fit(X_train, y_train)
    print(); print(model)
    # make predictions
    expected_y  = y_test
    predicted_y = model.predict(X_test)
    # summarize the fit of the model
    print(); print('naive_bayes.BernoulliNB(): ')
    print(); print(metrics.classification_report(expected_y, predicted_y, 
                   target_names=dataset.target_names))
    print(); print(metrics.confusion_matrix(expected_y, predicted_y))

    # fit a naive_bayes.GaussianNB() model to the data
    model = naive_bayes.GaussianNB()
    model.fit(X_train, y_train)
    print(); print(model)
    # make predictions
    expected_y  = y_test
    predicted_y = model.predict(X_test)
    # summarize the fit of the model
    print(); print('naive_bayes.GaussianNB(): ')
    print(); print(metrics.classification_report(expected_y, predicted_y, 
                   target_names=dataset.target_names))
    print(); print(metrics.confusion_matrix(expected_y, predicted_y))
    
    # fit a naive_bayes.MultinomialNB() model to the data
    model = naive_bayes.MultinomialNB()
    model.fit(X_train, y_train)
    print(); print(model)
    # make predictions
    expected_y  = y_test
    predicted_y = model.predict(X_test)
    # summarize the fit of the model
    print(); print('naive_bayes.MultinomialNB(): ')
    print(); print(metrics.classification_report(expected_y, predicted_y, 
                   target_names=dataset.target_names))
    print(); print(metrics.confusion_matrix(expected_y, predicted_y))    
Snippet_173()
```
