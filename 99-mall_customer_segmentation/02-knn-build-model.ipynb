```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
customers = pd.read_csv('Mall_Customers_Cluster.csv')
```

```python
customers
```

```python
data = pd.DataFrame(customers)
target = data['Cluster'].values
target
```

```python
train = data[['Age', 'Gender', 'Annual Income (k$)', 'Spending Score (1-100)']].values
train
```

```python
X_train, X_test, y_train, y_test = train_test_split(train, target, test_size=0.3, random_state=1) # 70% training and 30% test
```

```python
from sklearn.neighbors import KNeighborsClassifier

#Create KNN Classifier
knn = KNeighborsClassifier(n_neighbors=10)

print(X_test)

#Train the model using the training sets
knn.fit(X_train, y_train)

#Predict the response for test dataset
y_pred = knn.predict(X_test)
```

```python
from sklearn import metrics
print("Accuracy (k=10):",metrics.accuracy_score(y_test, y_pred))
```

```python
# save the model to disk using Pickle
import pickle
filename = 'knn_model.obj'
pickle.dump(knn, open(filename, 'wb'))
```
